#### Hypothesis testing with t-test/Mann-Whitney U test

The difference between expression protein abundance on the normal and tumor sample of the same patients.

```python
from scipy.stats import mannwhitneyu
```

```python
log_fold_list = []
log_pval_list = []

protein_list = []

i = 0; fc = None
for protein in proteomics_normal.columns:
    
    
    tumor = proteomics_tumor[protein].values
    normal = proteomics_normal[protein].values
    
    try:
        lgfc = np.mean(normal) - np.mean(tumor)
        #lgfc = math.log2(fc)
        _, p = mannwhitneyu(normal, tumor, method="exact")
        lgpval = -math.log10(p)
    except ValueError:
        continue
    
    if p == 0:
        p = 1e-5
    if fc == 0:
        fc = 1e-5
    
    log_fold_list.append(lgfc)
    log_pval_list.append(lgpval)
    protein_list.append(protein)
```

```python
df = pd.DataFrame(data={"Protein": protein_list, "LGFoldChange": log_fold_list, "LGpval": log_pval_list})
df["pval"] = df.LGpval.apply(lambda x: 1.0/math.exp(x))
df["Protein"] = df.Protein.apply(lambda x: x[0])
df["Color"] = df.apply(lambda row: 'red' if abs(row["LGFoldChange"]) > 2 and row["LGpval"] > 1.301 else 'blue', axis=1)
df["Color"] = df.apply(lambda row: 'green' if row["Protein"] == "IDH1" else row["Color"], axis=1)
df.head()
```
